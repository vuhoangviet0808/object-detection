```python
import fiftyone as fo
import fiftyone.zoo as foz

splits = ['train', 'test', 'validation']
for split in splits:
  dataset = foz.load_zoo_dataset(
      "open-images-v7",
      split=split,
      label_types=["detections"],
      classes=["Alpaca"],
      max_samples=1000,
      dataset_name="open-images-alpaca"
  )
```

```python
import os
import pandas as pd
from tqdm import tqdm
from PIL import Image

alpaca_id = '/m/0pcr'

output_dir = '/root/fiftyone/open-images-v7/train'
images_dir = os.path.join(output_dir, 'data')
labels_dir = os.path.join(output_dir, 'labels')

os.makedirs(labels_dir, exist_ok=True)
csv_file_path = os.path.join(labels_dir, 'detections.csv')
detections_df = pd.read_csv(csv_file_path)
detections_df = detections_df[detections_df['LabelName'] == alpaca_id]

image_filenames = detections_df['ImageID'].unique()

for img_filename in tqdm(image_filenames):
    detections = detections_df[(detections_df['ImageID'] == img_filename)]

    # Lấy kích thước ảnh từ metadata
    img_path = os.path.join(images_dir, img_filename)
    # with Image.open(img_path) as img:
    #     img_width, img_height = img.size

    filename_without_extension = os.path.splitext(img_filename)[0]

    label_filename = os.path.join(labels_dir, filename_without_extension + '.txt')

    with open(label_filename, 'w') as label_file:
        first_line = True  # Cờ để kiểm tra dòng đầu tiên
        for _, row in detections.iterrows():
            xmin, ymin, xmax, ymax = row['XMin'], row['YMin'], row['XMax'], row['YMax']

            # Tính toán các giá trị chuẩn YOLO
            x_center = (xmin + xmax) / 2
            y_center = (ymin + ymax) / 2
            width = xmax - xmin
            height = ymax - ymin

            # Chuẩn hóa các giá trị theo kích thước ảnh
            x_center_normalized = x_center  # / img_width
            y_center_normalized = y_center  # / img_height
            width_normalized = width  # / img_width
            height_normalized = height  # / img_height

            class_id = 0

            # Nếu không phải dòng đầu tiên, thêm dấu xuống dòng
            if not first_line:
                label_file.write("\n")

            # Ghi dòng vào tệp
            label_file.write(f"{class_id} {x_center_normalized} {y_center_normalized} {width_normalized} {height_normalized}")

            first_line = False
os.remove(csv_file_path)

```
